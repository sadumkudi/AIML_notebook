### Naive Bayes classfier
- Diabetes dataset

```python
import numpy as np
import pandas as pd
```

```python
# Load the dataset 
df = pd.read_csv('diabetes_dataset.csv')
```

```python
# Print the first few records of the dataset
df.head() 
```

```python
df.shape
```

```python
# Print the columns and datatype information of the dataset
df.info()
```

```python
df
```

```python
# Zero values are replaced with the mean values for these columns
Zero_not_acepted = ["Pregnancies", "Glucose","BloodPressure","SkinThickness","Insulin","BMI",
                    "DiabetesPedigreeFunction","Age"]

for column in Zero_not_acepted:
    df[column] = df[column].replace(0, np.NaN)
    mean = int(df[column].mean(skipna=True))
    df[column] = df[column].replace(np.NaN, mean)
    
    
df
```

```python
df.head()
```

```python
# Split the dataset into train data and test data
from sklearn.model_selection import train_test_split

X = df.iloc[:, 0:8]
y = df.iloc[:,8]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

#### Different types of Naive Bayes' Classfiers
1. Multinomial Naive Bayes
2. Bernoulli Naive Bayes
3. Gaussian Naive bayes


##### Multinomial Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB

# Creata a multinomial clasifier 
MultiNomialModel = MultinomialNB()

# Traing the model using the training set
MultiNomialModel.fit(X_train, y_train)

# Predict on test data
y_pred = MultiNomialModel.predict(X_test)
```

```python
# import confusion_matrix
from sklearn.metrics import confusion_matrix

# Evaluate Results
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
# Accuracy score
from sklearn.metrics import accuracy_score

print('Multinomial Accuracy =', accuracy_score(y_test, y_pred))
```

#### Bernoulli Naive Bayes

```python
# Import Bernoulli model
from sklearn.naive_bayes import BernoulliNB

# Create the Bernoulli classifier model object
BernoulliModel = BernoulliNB()

# Train the model using the training data set
BernoulliModel.fit(X_train, y_train)

# Predcict on test data
y_pred = BernoulliModel.predict(X_test)

# Evaluate Results
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
# Find the accuracy score
from sklearn.metrics import accuracy_score

print('Bernoulli Accuracy =', accuracy_score(y_test, y_pred))
```

#### Gaussian Naive bayes

```python
# Import Gaussian naive bayes' model 
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, accuracy_score

GaussianModel = GaussianNB()

GaussianModel.fit(X_train, y_train)

y_pred = GaussianModel.predict(X_test)

# Evaluate result


confusion_matrix(y_test, y_pred)
```

```python
print('Gaussian Accuracy =', accuracy_score(y_test, y_pred))
```

Gaussian Naive Bayes model gives a better accuracy score
